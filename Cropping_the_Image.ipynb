**Cropping the Image using CV2**

```python
image = plt.imread("20210806_133847_test4_R.bmp")
plt.imshow(image)
plt.show()
```

```python
patch = (100, 100)
center = (0, 0)
cropped_image = cv2.getRectSubPix(image,patch, center)
plt.imshow(cropped_image)
plt.show()
```

**Crop Function**

```python
def crop(image, center, width, height):
    return cv2.getRectSubPix(image, (width, height), center)
```

```python
cropped = crop(image, center=(190,220), width=80, height=90)
plt.imshow(cropped)
plt.show()
```

```python
cropping = crop(cropped, center=(10,10), width=100, height=160)
plt.imshow(cropping)
plt.show()
```
